```python
import pandas as pd
```

```python
flights = pd.read_csv("flights.csv")
flights.head()
```

```python
flights.isna().sum()
flights_clean = flights.dropna(axis = 0)
```

```python
flights_clean["dep_delay"] = flights_clean["dep_delay"].apply(lambda x: max(0, x))
flights_clean["arr_delay"] = flights_clean["arr_delay"].apply(lambda x: max(0, x))
```

```python
flights_clean
```

### 1. How many flights were recorded for each month in the dataset? And what is the average departure delay time for each month?
+ in the result box
+ in the result box

```python
flight_monthly = flights_clean['month'].value_counts()
flight_monthly
```

```python
flight_monthly_delay = flights_clean.groupby('month')['dep_delay'].mean()
flight_monthly_delay
```

### 2. Which were the top 5 most common destinations from the 'JFK' airport, based on the frequency

LAX, SFO, MIA,BOS, SJU

```python
flights_FromJFK = flights_clean[flights_clean['origin'] == "JFK"]
flights_FromJFK = flights_FromJFK['dest'].value_counts()
flights_FromJFK.head()
```

### 3. Calculate the total distance traveled by each carrier. Which carrier traveled the most distance overall?
United Airlines

```python
flight_distances_carrier = flights_clean.groupby('carrier')['distance'].sum()
flight_distances_carrier.sort_values(ascending=False)
```

### 4. How much departure delay do I need to expect when I use AA in December?(based on average)
13.93

```python
flights_delay = flights_clean[flights_clean['month'] == 12].groupby('carrier')['dep_delay'].mean()
print(flights_delay[flights_delay.index == "AA"])
```

### 5. Determine the top 5 busiest days (based on the number of flights). Report the year, month, day, and the number of flights.

```python
Top5 = flights_clean.groupby(['year','month','day']).size()
Top5 = pd.DataFrame(Top5, columns= ['Number of Flights'])
Top5.reset_index(inplace=True)
Top5.sort_values(by = "Number of Flights", ascending= False).head(5)
```
